## Task 1 - Predictive Modeling

```python
input_features=df[['Average Cost for two',
                   'Has Table booking',
                   'Has Online delivery',
                   'Is delivering now',
                   'Switch to order menu',
                   'Price range',
                   
                 ]]
input_features
```

```python
output_feature=df['Aggregate rating']
output_feature
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input_features,output_feature,test_size=0.3,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_model=LinearRegression()
lr_model.fit(x_train,y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor

dt_model=DecisionTreeRegressor(random_state=10)
dt_model.fit(x_train, y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model=RandomForestRegressor(n_estimators=100, random_state=10)
rf_model.fit(x_train, y_train)
```

```python
lr_pred=lr_model.predict(x_test)
dt_pred=dt_model.predict(x_test)
rf_pred=rf_model.predict(x_test)
```

```python
lr_accuracy=int(round(lr_model.score(x_test,y_test)*100))
dt_accuracy=int(round(dt_model.score(x_test,y_test)*100))
rf_accuracy=int(round(rf_model.score(x_test,y_test)*100))
```

```python
print("Linear Regression model accuracy : ",lr_accuracy,"%")
print("Decision tree model accuracy     : ",dt_accuracy,"%")
print("Random Forest model accuracy     : ",rf_accuracy,"%")
```
